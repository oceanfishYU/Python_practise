##panda的修改

```python
mydic = {'社團':['電影社','體育社','程式社','釣魚社','熱舞社','烹飪社'],'人數':[46,8,12,12,6,58],'入會費':[1500,2000,1000,1000,500,3000]}
mypd = pd.DataFrame(mydic)
mypd
```

```python
mypd.iloc[0]=['數學社',20,2000]
mypd
```

把第0列的資料改成數學社....

```python
mypd.loc[1]=['數學社',20,2000]
mypd
```

把第0列的資料改成數學社...

```python
mypd.社團 = mypd.社團.map({'數學社':0,'程式社':1,'釣魚社':2,'熱舞社':3,'烹飪社':4})
mypd
```

把社團那欄內使用map函數將數學社變成0，程式社變成1，釣魚社變成2....(map函數內用的是dictionary型態，鍵+值)

```python
mypd.loc[mypd.人數<20,'人數']=0
mypd.loc[(mypd.人數>=20) & (mypd.人數<=30),'人數']=1
mypd.loc[mypd.人數>20,'人數']=2
mypd
```

1、人數那欄<20的，人數改成0，2、人數那欄>=20 & <=30的，人數改成1，人數那欄>30的，人數改成2

##panda的合併

###利用append

```python
import pandas as pd
mydic = {'社團':['電影社','體育社','程式社','釣魚社','熱舞社','烹飪社'],'人數':[46,8,12,12,6,58],'入會費':[1500,2000,1000,1000,500,3000]}
mypd = pd.DataFrame(mydic)
mypd
```

```python
mypd1=pd.DataFrame([['遊戲社',20,2000],['遊戲社',50,2000]],columns=['社團','人數','入會費'])
mypd1
```

先製作一個要合併到裡面的panda

```python
mypd.append(mypd1,ignore_index=True)
```

使用append把mypd1合併到mypd裡，ignore_index=True意思是忽視原本mypd1的index。append不能新增列表!所以要先把資料用城panda。且此時mypd並沒有被改成這樣，要改成這樣的話前面要加mypd=....

###利用concat

```python
mylist1=[["電影社",46],["體育社",8],["程式社",12]]
mylist2=[["釣魚社",12],["熱舞社",6],["烹飪社",58]]
import pandas as pd
mypd1=pd.DataFrame(mylist1,columns=["社團","人數"])
mypd2=pd.DataFrame(mylist2,columns=["社團","人數"])
print(mypd1)
print(mypd2)
```

```python
pd.concat([mypd1,mypd2],ignore_index=True)
```

利用concat把pd1&pd2合併在一起，concat還可以做交集和連集(就是當欄位不同時能只顯示一樣的或都顯示但會有Non)，append不行

###利用merge合併

```python
mylist1=[["電影社",46],["體育社",8],["程式社",12],["釣魚社",12],["熱舞社",6],["烹飪社",58]]
mylist2=[["體育社",2000],["釣魚社",1000],["電影社",1500],["熱舞社",500],["程式社",1000],] #少了烹飪社
import pandas as pd
mypd1=pd.DataFrame(mylist1,columns=["社團","人數"])
mypd2=pd.DataFrame(mylist2,columns=["社團","入會費"])
print(mypd1)
print(mypd2)
```

```python
pd.merge(mypd1,mypd2)
```

利用merge把mypd1&mypd2合併，他會找到欄位一樣的(社團)且名稱一樣的，把兩筆資料做合併，烹飪社因為mypd2沒有，所以就沒有出現

mypd中人數沒有資料的為True

```python
mypd.dropna(axis=0)
```

mypd中有出現NaN的列刪除

```python
mypd.dropna(axis=1)
```

mypd中有出現NaN的欄刪除

```python
mypd.fillna(0)
```

mypd中Nan的裡面填入0

```python
mypd.fillna(mypd.mean())
```

```python
mypd.fillna(mypd.median())
```
